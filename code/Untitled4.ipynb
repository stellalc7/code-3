```python
import pandas as pd
import numpy as np
from __future__ import division
import calendar
import time
import datetime
from datetime import date, timedelta
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy import stats
from dateutil.parser import parse
from pandas.tslib import Timestamp
colorz=sns.color_palette("husl", 8)
#sns.palplot(sns.color_palette("husl", 8))

aqi=pd.read_excel('/Users/stella/Desktop/realeco_sn0135-ldsa_2017-04-01t040000_2017-05-24t035900.xlsx',delimiter=',',skiprows=0)
aqu=pd.read_csv('/Users/stella/Desktop/1329_PEGASOR A_20170524182426.csv',delimiter=',',skiprows=4)

aqu=aqu[['DateTime','MEASURE_PA_LABEL','MEASURE_PA_LABEL.1','MEASURE_PA_LABEL.2']]
list(aqu.columns)
aqu.columns=['DateTime','ldsa1','ldsa2','ldsa3']
aqu['DateTime']=aqu['DateTime'].apply(lambda x: parse(x))
```

```python
dt1=timedelta(hours=1)
```

```python
aqi['time']=aqi['time']-dt1
```

```python
hrs=aqi['time'].apply(lambda x: x.hour)
```

```python
aqi=aqi.set_index('time')
aqu=aqu.set_index('DateTime')
df=aqu.join(aqi)

df=df.reset_index()
df=df.dropna()

df=df.rename(columns={'index':'datetime'})
df = df.where(df >= 0, np.nan)
```

```python
df['u_avg']=df[['ldsa1','ldsa3']].mean(axis=1)
```

```python
plt.figure(figsize=(12,6))
#plt.plot(df['datetime'],df['ldsa1'])
#plt.plot(df['datetime'],df['ldsa2'])
#plt.plot(df['datetime'],df['ldsa3'])
plt.plot(df['datetime'],df['u_avg'])
plt.plot(df['datetime'],df['value (µm²/cm³)'])
plt.legend()
```
